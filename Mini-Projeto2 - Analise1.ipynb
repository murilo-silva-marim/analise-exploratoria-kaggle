## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
copiaDF = df.copy()
plt.hist(copiaDF['yearOfRegistration'], 60)
plt.xlabel("Ano de Registro", fontsize=20)
plt.ylabel("Densidade (KDE)", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title("Distribuição de Veículos com base no Ano de Registro", fontsize=20)

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

# Salvando o plot
plt.savefig("plots/Analise1/vehicle-distribution.png")

plt.show()
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
copiaDF.head()
veiculos = copiaDF['vehicleType'].unique()
columns = [copiaDF[copiaDF['vehicleType'] == x]['price'] for x in veiculos]
box = plt.boxplot(columns, patch_artist=True)
plt.xticks(np.arange(len(veiculos)) + 1, veiculos, fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Tipo de Veículo", fontsize=20)
plt.ylabel("Range de Preço", fontsize=20)
plt.title("Análise de Outliers", fontsize=20)

colors = ['#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF',
          '#415b81', '#23562c',
          '#fff12f', '#979583',
          '#6248af']

for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)    

# Salvando o plot
plt.savefig("plots/Analise1/price-vehicleType-boxplot.png")

plt.show()
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria
norm = plt.Normalize(0, copiaDF['vehicleType'].value_counts().values.max())
colors = plt.cm.Purples(norm(copiaDF['vehicleType'].value_counts().values))
copiaDF['vehicleType'].value_counts().plot(kind = 'bar', color=colors)
for index,data in enumerate(copiaDF['vehicleType'].value_counts()):
    plt.text(x=index - .18, y =data+1000 , s=f"{data}" , fontsize=15)
plt.xticks(rotation=0, fontsize=15)
plt.yticks(rotation=0, fontsize=15)
plt.xlabel("Tipo de Veículo", fontsize=20)
plt.ylabel("Total de Veículos Para Venda", fontsize=20)
plt.title("Contagem total de veículos à venda conforme o tipo de veículo", fontsize=20)

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

# Salvando o plot
plt.savefig("plots/Analise1/count-vehicleType.png")

plt.show()
```
