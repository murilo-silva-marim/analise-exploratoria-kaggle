```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
copiaDF = df.copy()
base = copiaDF['brand'].value_counts().sort_values()
norm = plt.Normalize(0, base.values.max())
colors = plt.cm.Reds(norm(base.values))
g, ax = plt.subplots()
ax.barh(base.index, base.values, color=colors)
g.set_size_inches(18.5, 10.5, forward=True)
plt.xlabel("Número de Veículos", fontsize=20)
plt.ylabel("Marca", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=14)
plt.title("Veículos por Marca", fontsize=20)
plt.show()
```

```python
# Salvando o plot
g.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
widthBar = .2
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5, forward=True)
valoresGrafico = copiaDF.groupby(['gearbox', 'vehicleType']).mean()['price'].reset_index()
i = 0
for gearbox in valoresGrafico['gearbox'].unique():
    y = valoresGrafico[valoresGrafico['gearbox'] == gearbox]['price']
    x = np.arange(len(y)) + i
    ax.bar(x, y, widthBar, label=gearbox)
    i = i + .2
    
plt.xticks(np.arange(len(valoresGrafico['vehicleType'].unique())) + (i / len(valoresGrafico['gearbox'].unique())), valoresGrafico['vehicleType'].unique(), fontsize=15)
plt.yticks(fontsize=15)
plt.legend(title='gearbox', fontsize=15)
plt.xlabel("Tipo de Veículo", fontsize=20)
plt.ylabel("Preço Médio", fontsize=20)
plt.title("Preço médio dos veículos por tipo de veículo e tipo de caixa de câmbio", fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
