```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
copiaDF = df.copy()
widthBar = .2
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5, forward=True)
valoresGrafico = copiaDF.groupby(['gearbox', 'fuelType']).mean()['price'].reset_index()
i = 0
for gearbox in valoresGrafico['gearbox'].unique():
    y = valoresGrafico[valoresGrafico['gearbox'] == gearbox]['price']
    x = np.arange(len(y)) + i
    ax.bar(x, y, widthBar, label=gearbox)
    i = i + .2
    
plt.xticks(np.arange(len(valoresGrafico['fuelType'].unique())) + (i / len(valoresGrafico['gearbox'].unique())), valoresGrafico['fuelType'].unique(), fontsize=15)
plt.yticks(fontsize=15)
plt.legend(title='gearbox', fontsize=15)
plt.xlabel("Tipo de Combustível", fontsize=20)
plt.ylabel("Preço Médio", fontsize=20)
plt.title("Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio", fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
widthBar = .2
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5, forward=True)
valoresGrafico = copiaDF.groupby(['gearbox', 'vehicleType']).mean()['powerPS'].reset_index()
i = 0
for gearbox in valoresGrafico['gearbox'].unique():
    y = valoresGrafico[valoresGrafico['gearbox'] == gearbox]['powerPS']
    x = np.arange(len(y)) + i
    ax.bar(x, y, widthBar, label=gearbox)
    i = i + .2
    
plt.xticks(np.arange(len(valoresGrafico['vehicleType'].unique())) + (i / len(valoresGrafico['gearbox'].unique())), valoresGrafico['vehicleType'].unique(), fontsize=15)
plt.yticks(fontsize=15)
plt.legend(title='gearbox', fontsize=15)
plt.xlabel("Tipo de Veículo", fontsize=20)
plt.ylabel("Potência Média", fontsize=20)
plt.title("Potência Média de um veículo por tipo de veículo e tipo de caixa de câmbio", fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
