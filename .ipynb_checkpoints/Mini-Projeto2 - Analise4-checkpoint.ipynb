## Análise 4

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
# Calcule a média de preço por marca e por veículo
copiaDF = df.copy()
resultadoDF = copiaDF.groupby(['brand', 'vehicleType']).mean()['price'].reset_index()
resultadoDF = resultadoDF.rename(columns={'price': 'avgPrice'})
resultadoDF.head()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo
pivotDF = pd.pivot_table (resultadoDF, values = "avgPrice", 
                            index = "brand", columns = "vehicleType")
pivotDF = pivotDF.fillna(0)
pivotDF = pivotDF.round(decimals=0)
fig, ax = plt.subplots(figsize=(15, 20))
sns.heatmap(pivotDF, cmap="YlGnBu", annot=True, fmt='g')
plt.xlabel("Tipo de Veículo", fontsize=20)
plt.ylabel("Marca", fontsize=20)
plt.title("Heatmap - Preço médio de um veículo por marca e tipo de veículo", fontsize=20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise4/heatmap-price-brand-vehicleType.png")
```
